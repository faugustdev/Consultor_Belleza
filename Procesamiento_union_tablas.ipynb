```python
import pandas as pd
```

```python
data_cosmetics = pd.read_csv('cosmetics_final.csv')
df_form = pd.read_csv('df_form.csv')
```

```python
data_cosmetics.head(2)
```

```python
print(data_cosmetics.isnull().sum())
```

```python
data_cosmetics = data_cosmetics.drop(['Dry','Normal','Oily','Sensitive'], axis = 1)
```

```python
data_cosmetics.head(2)
```

```python
data_cosmetics['Label'].value_counts()
```

```python
# Defino un diccionario de mapeo
label_mapping = {
    'Moisturizer': 1,
    'Cleanser': 2,
    'Face Mask': 3,
    'Treatment': 4,
    'Eye cream': 5,
    'Sun protect': 6
}
```

```python
# Utilizo la función map para asignar los valores numéricos a la nueva columna 'Label_Num'
data_cosmetics['Label_Num'] = data_cosmetics['Label'].map(label_mapping)
```

```python
data_cosmetics.head(1)
```

```python
data_cosmetics = data_cosmetics.drop(['Label'], axis = 1)
```

```python
data_cosmetics.head(1)
```

```python
data_cosmetics.shape
```

```python
df_form.head(2)
```

```python
df_form.shape
```

```python
print(data_cosmetics.isnull().sum())
```

```python
print(df_form.isnull().sum())
```

```python
df_form.head(1)
```

```python
data_cosmetics_form = pd.concat([data_cosmetics,df_form])
```

```python
data_cosmetics_form.shape
```

```python
data_cosmetics_form.info()
```

```python
print(data_cosmetics_form.isnull().sum())
```

```python
data_cosmetics_form_filled = data_cosmetics_form.fillna(data_cosmetics_form.median())
```

```python
print(data_cosmetics_form_filled.isnull().sum())
```

```python
data_cosmetics_form_filled.to_csv('df_cosmetics_form.csv', index = False)
```
